```python
#!/users/jflournoy/.conda/envs/abcd_ml_3.7/bin/python
print('importing libs...')
import numpy as np
import pandas as pd
import argparse
from joblib import dump, load
import logging
from pathlib import Path

listing = [x for x in Path('out/').rglob('*.pkl')]
filesdf = pd.DataFrame([str(x.stem).split('_') for x in listing],
            columns = ['script', 'predictors', 'outcome', 'summary', 'time', 'cv_iter', 'slurmid', 'arrayid'])
filesdf['fn'] = listing

scores_df = pd.DataFrame(columns = ['score_test', 'score_train'])
for index, row in filesdf.iterrows():
    print("Reading {}".format(row.fn))
    with open(row.fn, 'rb') as f:
        out_dict = load(f)
    scores_df = scores_df.append(pd.DataFrame({key: [out_dict[key]] for key in ['score_test', 'score_train']}))
    test_data_df = pd.DataFrame({key: list(np.concatenate(out_dict[key]).flat) for key in ['Y_test', 'Y_pred_test']})
    for i in row.index:
        test_data_df[i] = row[i]

scores_df = pd.concat([filesdf, scores_df.reset_index()], axis = 1)
scores_df.to_csv('model_train_test_scores.csv', index = False)
test_data_df.to_csv('model_test_data.csv', index = False)
```
